# FIND MINIMUM,MAXIMUM AND AVERGE BASEPAY

```python
data['BasePay'] = pd.to_numeric(data['BasePay'], errors='coerce')

# Drop rows with NaN values in 'BasePay' column
data.dropna(subset=['BasePay'], inplace=True)

maximum=data['BasePay'].max()
minimum=data['BasePay'].min()
Average=data['BasePay'].mean()
print("Maximum number",maximum)
print("Minimum number",minimum)
print("Average number",Average)
```

# HOW MUCH ALBERT PARDINI MAKE (INCULDE DENEFITS)?

```python
import math
albert_data = data[data['EmployeeName'] == 'ALBERT PARDINI']

if not albert_data.empty:
    base_pay = albert_data['BasePay'].iloc[0]
    overtime_pay = albert_data['OvertimePay'].iloc[0]
    other_pay = albert_data['OtherPay'].iloc[0]
    benefits = albert_data['Benefits'].iloc[0]
    if math.isnan(benefits):
        benefits = 0

    total_compensation = base_pay + overtime_pay + other_pay + benefits

    print(f'Total Compensation ${total_compensation}')
```
